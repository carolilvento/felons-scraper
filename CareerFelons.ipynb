```python
import requests
from bs4 import BeautifulSoup
from dateutil.parser import parse
from nameparser import HumanName
import re
import csv

with open('felons.csv', mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    url = "http://www.fdle.state.fl.us/coflyer/default.asp"
    cookies = {}
    for i in range(1,2):
        pagenum = 1
        data = {
            "cmd": "Search", 
            "dhtmlType": "ie", 
            "selPage": i, 
            "PageNumber": i, 
            "PageSource": 0, 
            "imgSubmit.x": 26,
            "imgSubmit.y": 7,
            "SelectChks": 0
        }

        page = requests.post(url, data=data, headers=cookies)
        soup = BeautifulSoup(page.content)

        table = soup.find('table')
        table = table.findAll('table')[4]
        row = table.findAll('tr')
        for x in row[1:]:
        #     print(x)
            subjecturl = x.find('a')
            name = subjecturl.text
            name = HumanName(name)
            urlx = x.find('a').attrs['href']
            urlx = 'http://www.fdle.state.fl.us/coflyer/' + urlx
            td = x.findAll('td')[1]
            res = parse(td.text, fuzzy=True)
            dob = str(res)[:10]
            #print(name.last,name.first,name.middle,urlx,dob)
            pagenum = i + 1
        cookies = page.cookies.get_dict()
        for i in range(2,866):

            data = {
                "cmd": "Page", 
                "dhtmlType": "ie", 
                "selPage": i, 
                "PageNumber": i, 
                "PageSource": 0, 
                "SelectChks": 0
            }

            #print(cookies)

            page = requests.post(url, data=data, cookies=cookies)
            #print(data)
    #         cookies = page.cookies.get_dict()
            soup = BeautifulSoup(page.content)

            table = soup.find('table')
            table = table.findAll('table')[4]
            row = table.findAll('tr')
            for x in row[1:]:
            #     print(x)
                subjecturl = x.find('a')
                name = subjecturl.text
                name = HumanName(name)
                urlx = x.find('a').attrs['href']
                urlx = 'http://www.fdle.state.fl.us/coflyer/' + urlx
                td = x.findAll('td')[1]
                #print(td.text)
                match = re.search(r'\d+/\d+/\d{4}', td.text)
                #print(match)
                #print(match.group(0))
                try:
                    res = parse(match.group(0), fuzzy=True)
                except:
                    res = '2022-02-02 00:00:00'
                #print(res)
                dob = str(res)[:10]
                row = [name.last,name.first,name.middle,urlx,dob]
                print(row)
                csv_writer.writerow(row)


```
